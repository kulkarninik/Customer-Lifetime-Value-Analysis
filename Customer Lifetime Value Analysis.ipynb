# Objectives

- Create a visualization for the revenue generated by the customer.
- Compare the cost of acquisition across various channels and determine which ones are the most and least profitbale
- find out which channels are most and least effective at converting customers?
- Calculate the total revenue by channel and analyze the most and least profit channels in terms of generating revenue.

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
pio.templates.default = "plotly_white"
```

```python
data = pd.read_csv("C:/Users/Nikita/Desktop/IScale Notes/Files/customer_acquisition_data.csv")
data.head()
```

# Objective 1: Make a Visualization for the distribution of Customer Acquisition Cost using Histograms

```python
fig = px.histogram(data,
                  x = "cost",
                  nbins = 20,
                  title = "Distribution of Acquistion Cost")
fig.show()
```

# Objective 2: Create a visualization for the revenue generated by the customer.

```python
data.head()
```

```python
fig = px.histogram(data,
                  x = "revenue",
                  nbins = 20,
                  title = "Distribution of Revenue")
fig.show()
```

# Objective 3: Compare the cost of acquisition across various channels and determine which ones are the most and least profitbale

```python
data.head()
```

```python
cost_by_channel = data.groupby('channel')['cost'].mean().reset_index()

fig = px.bar(cost_by_channel,
                  x = 'channel',
                  y = 'cost',
            title = 'Customer Acquistion Cost by Channel', width =600, height = 600)

fig.show()
```

# Objective 4 : find out which channels are most and least effective at converting customers?

```python
data.head()
```

```python
conversation_by_channel = data.groupby('channel')['conversion_rate'].mean().reset_index()

fig = px.bar(conversation_by_channel,
            x = 'channel',
            y = 'conversion_rate',
            title = 'Most and Least Effective Channel', width =1000, height = 400)

fig.show()
```

# Objective 5: Calculate the total revenue by channel and analyze the most and least profit channels in terms of generating revenue.

```python
data.head()
```

```python
revenue_by_channel = data.groupby('channel')['revenue'].sum().reset_index()

fig = px.pie(revenue_by_channel,
            values = 'revenue',
            names = 'channel',
            title = 'Total Revenue by Channel',
             hole = 0.5
            )

fig.show()
```

# Objective 6 :calculate the return on inverstment (ROI) for each channel

```python
data.head()
```

```python
data['roi'] = data['revenue']/data['cost']

roi_by_channel = data.groupby('channel')['roi'].mean().reset_index()

fig = px.bar(roi_by_channel,
            x = 'channel',
            y = 'roi',
            title= 'Return on Investment (ROI) by Channel')

fig.show()
```
